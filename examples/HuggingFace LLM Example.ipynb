```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

texts = [
    "This molecule is toxic",
    "This compound is safe",
    "This drug has harmful side effects"
]

results = classifier(texts)

for text, result in zip(texts, results):
    print(f"Input: {text}")
    print(f"Output: {result}\n")
```

## Explanation

This notebook demonstrates how HuggingFace Transformer models can be used for natural language processing tasks within a DeepChem-like workflow.

### What is happening here?

- We use a pretrained transformer model for **sentiment classification**.
- Input texts are passed as a batch to the model.
- The model returns predictions (e.g., positive/negative sentiment with confidence scores).

### Why is this relevant to DeepChem?

Although this example uses simple text inputs, the same approach can be extended to:

- Molecular property prediction (e.g., toxicity, solubility)
- Drug discovery tasks using textual or sequence-based representations
- Integration of large language models (LLMs) into scientific machine learning pipelines

### Key Takeaways

- HuggingFace pipelines make it easy to use powerful LLMs
- Batch processing improves efficiency
- This approach can be integrated with DeepChem models for advanced applications
