## **Introduction**  <a name="Introduction"></a>

Protein structure refers to the arrangement of atoms within a protein. Proteins are complex molecules made up of chains of amino acids, and their structures determine how they perform their specific biological tasks. The structure of a protein is intrinsically linked to its function; even minor changes in the structure can significantly impact the protein's function and, by extension, can affect biological processes and health.

Protein structure prediction is the inference of the three-dimensional structure of a protein from its amino acid sequenceâ€”that is, the prediction of its secondary and tertiary structure from primary structure. Protein structure prediction is one of the most important goals pursued by computational biology; it is important in medicine (for example, in drug design) and biotechnology (for example, in the design of novel enzymes). We'll learn more about diffrent structure of proteins, how to predict the structure and it's application in different task.


Protein structures are typically described at four different levels: primary, secondary, tertiary and quaternary structure.


### **Primary Structure** <a name="primary"></a>

The primary structure of a protein refers to its unique sequence of amino acids linked together by peptide bonds. This sequence is determined by the genetic code and dictates how the protein will fold into its three-dimensional structure, ultimately determining its function. Here are the key aspects of the primary structure:

* **Amino Acid Sequence:** The primary structure is a linear chain of amino acids. Each amino acid has a central carbon atom (alpha carbon) bonded to an amino group (-NH2), a carboxyl group (-COOH), a hydrogen atom, and a distinctive side chain (R group) that varies between different amino acids.

* **Peptide Bonds:** Amino acids are linked together by peptide bonds, which form through a dehydration reaction between the carboxyl group of one amino acid and the amino group of the next. This bond creates a repeating backbone of nitrogen-carbon-carbon atoms.

* **Directionality:** The sequence has directionality, typically written from the amino (N) terminus to the carboxyl (C) terminus. This orientation is critical for the proper folding and function of the protein.

* **Genetic Encoding:** The primary structure is encoded by the corresponding gene in the organism's DNA. The sequence of nucleotides in the gene determines the order of amino acids in the protein through the processes of transcription and translation.

The primary structure dictates how the protein will fold into secondary, tertiary, and quaternary structures. Specific sequences can lead to particular structural motifs like alpha helices and beta sheets.

### **Secondary Structure** <a name="secondary"></a>

The secondary structure of a protein refers to the local, repetitive folding patterns of the polypeptide chain, stabilized by hydrogen bonds between the backbone atoms. These structures contribute to the overall shape and stability of the protein. The two most common types of secondary structures are the alpha helix and the beta sheet, but there are other elements like turns and loops as well. The two most common secondary structural elements are alpha helices and beta sheets.

1. **Alpha Helix:**

- Structure: The alpha helix is a right-handed coil where the backbone of the polypeptide chain forms a spiral structure.
- Hydrogen Bonds: Hydrogen bonds form between the carbonyl oxygen of one amino acid and the amide hydrogen of another amino acid four residues earlier in the sequence. This bonding pattern stabilizes the helical structure.
- R Groups: The side chains (R groups) of the amino acids in an alpha helix extend outward from the helix, allowing them to interact with the environment or other parts of the protein.

Fig 2. alpha-helix structure of a protein([source](https://commons.wikimedia.org/wiki/File:Protein_structure_(2)-en.svg))


2. **Beta Sheet:**

- Structure: Beta sheets consist of beta strands connected laterally by at least two or three backbone hydrogen bonds, forming a sheet-like arrangement.
- Parallel and Antiparallel: Beta sheets can be parallel (strands run in the same direction) or antiparallel (strands run in opposite directions). The hydrogen bonding patterns differ slightly between these arrangements.
- R Groups: The side chains of amino acids in beta strands alternate between extending above and below the plane of the sheet, allowing for interactions with other molecules or structural elements.

### **Tertiary Structure** <a name="tertiary"></a>

The tertiary structure of a protein refers to the overall three-dimensional shape formed by the entire polypeptide chain, including the arrangement of its secondary structural elements (alpha helices, beta sheets, turns, and loops). This level of structure is stabilized by various interactions between the side chains (R groups) of the amino acids.

image.png

Fig 4. Terrtiay structure of protein. ([source](https://www.ebi.ac.uk/training/online/courses/biomacromolecular-structures/proteins/levels-of-protein-structure-primary/levels-of-protein-structure-tertiary/))
The tertiary structure is the complete 3D configuration of a single polypeptide chain, formed by the folding and coiling of the chain into a specific shape that is energetically favorable.

**Stabilizing Interactions in tertiary structure of protein**

Protein tertiary structure is stabilized by several interactions. Hydrogen bonds can form between side chains or between side chains and the backbone, enhancing structural stability. Ionic bonds (salt bridges) occur between positively and negatively charged side chains, such as lysine and aspartate, providing substantial electrostatic stabilization. Hydrophobic interactions drive hydrophobic side chains to cluster in the protein's interior, avoiding water and stabilizing the structure. Van der Waals forces are weak attractions between atoms in close proximity that further contribute to stability. Additionally, disulfide bridges, covalent bonds between cysteine residues, create strong stabilizing connections within the protein.


**Domains in tertiary structure of protein**

Domains are distinct, functional, and structural units within a protein's tertiary structure. Each domain typically folds independently into a stable, three-dimensional shape, often resembling small globular proteins. Domains can be independently stable and functional units within the protein

Domains play crucial roles in the protein's overall function. They can act as binding sites for other molecules, participate in the protein's catalytic activity, or mediate interactions with other proteins or cellular components. Because of their modular nature, domains can be found in different proteins, often conferring similar functions across diverse biological contexts. This modularity allows for the evolutionary recombination of domains, leading to the creation of new proteins with novel functions.



**Functional Sites in tertiary structure of protein**

The tertiary structure creates specific functional sites such as active sites, binding sites, and allosteric sites. These sites are formed and defined by the unique three-dimensional arrangement of amino acids within the protein.These sites are critical for the protein's biological activity, allowing it to interact specifically with other molecules, substrates, or ligands. Here are the main types of functional sites:

- **Active Sites:**
Active sites are regions on enzymes where substrates bind and undergo chemical reactions. They consist of specific amino acid residues that facilitate catalysis by lowering the activation energy of the reaction. It is where substrate molecules bind and are transformed into products through chemical reactions. The precise arrangement of amino acids in the active site ensures the enzyme's specificity and efficiency, making it essential for metabolic processes and biochemical reactions within the cell.



- **Binding Sites:**
Binding sites are regions where proteins interact with specific molecules (ligands), such as other proteins, DNA, RNA, or small molecules. These sites are formed by non-covalent interactions and are crucial for various cellular functions, including signal transduction, molecular recognition, and the regulation of biochemical pathways. Binding sites help maintain cellular homeostasis and facilitate communication within and between cells.


- **Allosteric Sites:**
Allosteric sites allow for the regulation of protein activity through the binding of effector molecules. When an effector binds to an allosteric site, it induces a conformational change that can either enhance or inhibit the protein's function. This regulation is vital for controlling metabolic pathways and ensuring that cellular processes are responsive to changing conditions and signals.

- **Interaction Domains:**
Interaction domains are specialized regions within proteins that mediate interactions with other proteins. These domains are essential for the assembly of multi-protein complexes, signal transduction pathways, and the formation of structural components within the cell. They enable proteins to perform coordinated functions and facilitate complex cellular activities.


- **DNA/RNA Binding Sites:**
DNA and RNA binding sites are critical for the regulation of gene expression and the processing of genetic information. Proteins that bind to DNA can control transcription, replication, and repair processes, while RNA-binding proteins are involved in splicing, transport, and translation of RNA. These interactions are fundamental to the accurate transmission and expression of genetic information.


- **Catalytic Triads and Dyads:**
Catalytic triads (or dyads) are arrangements of amino acids in enzyme active sites that are crucial for their catalytic function. These residues work together to facilitate the breakdown or synthesis of molecules, playing a central role in various biochemical reactions. Their significance lies in their ability to increase reaction rates and ensure the specificity and precision of enzymatic activities.





**Role of tertiary structure in function of protein**

The tertiary structure of a protein is fundamental to its function. The tertiary structure creates precise shapes and surfaces that allow proteins to interact specifically with other molecules, including substrates, ligands, and other proteins. This specificity is essential for the protein's ability to bind to its targets with high affinity and selectivity.Misfolding of the tertiary structure can lead to loss of function or gain of toxic function, contributing to diseases such as Alzheimer's, Parkinson's, and cystic fibrosis.

The tertiary structure also provides stability to the protein by folding into a compact, stable conformation that is resistant to denaturation under physiological conditions. It brings together amino acids from different parts of the polypeptide chain to form functional sites such as active sites, binding sites, and allosteric sites.These sites are critical for the protein's biological activity, enabling it to perform its specific functions, such as catalysis, signaling, and regulation.


The flexibility of the tertiary structure allows proteins to undergo conformational changes in response to binding events or environmental changes. These changes can modulate the protein's activity and interactions.
For instance, allosteric proteins change their conformation upon binding of an effector molecule at an allosteric site, which can enhance or inhibit their activity.


**Why do we predict protein structure**

As we saw above the 3D structure of protein is crucial for understanding it's function and interactions within biological systems. Protein structure prediction plays a crucial role in drug discovery for several reasons. Some of the reasons are:

1. Unveiling Druggable Pockets: Drugs work by binding to specific sites on a protein, often called pockets. Knowing the 3D structure of a protein allows researchers to identify these pockets and assess their suitability for drug binding. Tools like fpocket can analyze the predicted structure and provide metrics like pocket volume and hydrophobicity, which indicate how well a drug molecule might fit and interact.

2. Understanding Drug-Protein Interactions: Protein structure prediction helps visualize how a potential drug molecule interacts with the target protein at the atomic level. This allows researchers to analyze potential effects on protein function or stability. For instance, the simulation might reveal if the drug binding disrupts the protein's active site, potentially hindering its ability to perform its normal function.

3. Functional Domain prediction : Protein structure is critical in the prediction and understanding of functional domains. Functional domains are specific regions within a protein that are responsible for particular biological functions. Understanding the 3D structure of a protein helps in identifying these domains and predicting their functions based on several structural features.

4. Guiding Experimental Validation: Protein structure prediction helps prioritize drug candidates for further, expensive wet-lab experiments. By filtering out molecules with low predicted binding affinity or those that might disrupt protein function, researchers can focus their resources on the most promising candidates with a higher chance of success.


## **ESMFold** <a name="ESM"></a>

In  this tutorial we'll use ESMFold for predicting protein's tertiary structure.

ESMFold ([paper link](https://www.biorxiv.org/content/10.1101/2022.07.20.500902v2)) is a powerful tool in protein structure prediction, leveraging advanced deep learning techniques to predict the three-dimensional structures of proteins from their amino acid sequences. It is built on a 15 billion parameter Transformer language model called ESM-2 that learns the statistical patterns in protein sequences to infer their folded structures. It uses a learned language model(also known as LLM) representation that requires only the amino acid sequence as input, unlike other methods like AlphaFold2 which need additional databases.

ESMFold works by first encoding the input protein sequence using the ESM-2 language model, which learns representations capturing structural information. These representations are then fed into a downstream component similar to AlphaFold2 that predicts the 3D coordinates of the atoms in the folded structure.

ESMFold was trained using a large dataset of protein sequences from the UniRef90 database. All PDB chains until May 1, 2020 with resolution greater than 2Ã… were used as the training set, providing a diverse set of high-quality protein structures



**Let's get started!**

We'll use hugging face to import our model.

````python
#load the tokenizer and model from hugging face
from transformers import AutoTokenizer, EsmForProteinFolding

tokenizer = AutoTokenizer.from_pretrained("facebook/esmfold_v1")
model = EsmForProteinFolding.from_pretrained("facebook/esmfold_v1", low_cpu_mem_usage=True)

model = model.cuda()
````

 Since ESMFold is quite a large model, there are some considerations regarding memory usage and performance. This step is important if you have RAM less than 16GB or you are using free colab notebook.



````python
import torch
model.esm = model.esm.half()
torch.backends.cuda.matmul.allow_tf32 = True
model.trunk.set_chunk_size(64)
````

### **Folding Single Chain of Protein** <a name="sESM"></a>

Input Protein Sequence:



````python
#This is the sequence for human GNAT1
test_protein = "MGAGASAEEKHSRELEKKLKEDAEKDARTVKLLLLGAGESGKSTIVKQMKIIHQDGYSLEECLEFIAIIYGNTLQSILAIVRAMTTLNIQYGDSARQDDARKLMHMADTIEEGTMPKEMSDIIQRLWKDSGIQACFERASEYQLNDSAGYYLSDLERLVTPGYVPTEQDVLRSRVKTTGIIETQFSFKDLNFRMFDVGGQRSERKKWIHCFEGVTCIIFIAALSAYDMVLVEDDEVNRMHESLHLFNSICNHRYFATTSIVLFLNKKDVFFEKIKKAHLSICFPDYDGPNTYEDAGNYIKVQFLELNMRRDVKEIYSHMTCATDTQNVKFVFDAVTDIIIKENLKDCGLF"
````

Tokenize the protein sequence

````python
#tokenize the input protein
tokenized_input = tokenizer([test_protein], return_tensors="pt", add_special_tokens=False)['input_ids']

#If you're using a GPU, you'll need to move the tokenized data to the GPU now.
tokenized_input = tokenized_input.cuda()
````

Now, we predict the 3D structure. It's as simple as this.

````python
#generate 3d structure
with torch.no_grad():
    output = model(tokenized_input)
````

We save the predicted structure in a PDB file. It is important to do so to be able to use this structure for other task such as druggability assessment, functional domain prediction etc.

````python
from transformers.models.esm.openfold_utils.protein import to_pdb, Protein as OFProtein
from transformers.models.esm.openfold_utils.feats import atom14_to_atom37

def convert_outputs_to_pdb(outputs):
    final_atom_positions = atom14_to_atom37(outputs["positions"][-1], outputs)
    outputs = {k: v.to("cpu").numpy() for k, v in outputs.items()}
    final_atom_positions = final_atom_positions.cpu().numpy()
    final_atom_mask = outputs["atom37_atom_exists"]
    pdbs = []
    for i in range(outputs["aatype"].shape[0]):
        aa = outputs["aatype"][i]
        pred_pos = final_atom_positions[i]
        mask = final_atom_mask[i]
        resid = outputs["residue_index"][i] + 1
        pred = OFProtein(
            aatype=aa,
            atom_positions=pred_pos,
            atom_mask=mask,
            residue_index=resid,
            b_factors=outputs["plddt"][i],
            chain_index=outputs["chain_index"][i] if "chain_index" in outputs else None,
        )
        pdbs.append(to_pdb(pred))
    return pdbs
pdb = convert_outputs_to_pdb(output)
````

````python
# Save the PDB file
with open("./our_predicted_3D_tructure.pdb", "w") as f:
    f.write(pdb[0])
````

Visualise the Predicted Structure

````python
#visualise the protein structure
import py3Dmol
view = py3Dmol.view(js='https://3dmol.org/build/3Dmol.js', width=800, height=400)
view.addModel("".join(pdb), 'pdb')
view.setStyle({'model': -1}, {"cartoon": {'color': 'spectrum'}})
````

Looks good! We can colour it differently, though - our model outputs a `plddt` field containing probabilities for each atom, indicating how confident it is in that part of the structure. In the conversion function above we added the `plddt` field in the `b_factors` argument, so it was included in our `pdb` string. Let's use it so that we can see high- and low-confidence areas of the structure visually!

The pLDDT (predicted Local Distance Difference Test) score is a key output of the ESMFold protein structure prediction model. It provides a measure of the confidence in the predicted structure at the per-residue level.
It ranges from 0 to 1, with 1 indicating the highest confidence in the predicted structure for that residue.


````python
# The plddt field is scaled from 0-1 on version of ESMFold used in this notebook but will be updated
# to match AlphaFold's scale of 0-100 in future versions.
#Blue indicates high confidence

if torch.max(output['plddt']) <= 1.0:
    vmin = 0.5
    vmax = 0.95
else:
    vmin = 50
    vmax = 95

view.setStyle({'cartoon': {'colorscheme': {'prop':'b','gradient': 'roygb','min': vmin,'max': vmax}}})
````

### **Folding Multiple Chain** <a name="mESM"></a>

Many proteins exist as complexes, either as homopolymers (multiple copies of the same peptide) or heteropolymers (complexes of different peptides). To generate folds for such structures in ESMFold, we use a method from the research paper: inserting a flexible glycine residue "linker" between each chain to be folded simultaneously. Additionally, we offset the position IDs for each chain, making the model treat them as distant segments of a single long chain.

We'll use Glucosamine-6-phosphate deaminase (Uniprot: Q9CMF4) from the paper as an example.

First, we define the sequence of the monomer, and the poly-G linker we want to use. Then we stick two copies of the monomer together with the linker in between.

````python
sequence = "MRLIPLHNVDQVAKWSARYIVDRINQFQPTEARPFVLGLPTGGTPLKTYEALIELYKAGEVSFKHVVTFNMDEYVGLPKEHPESYHSFMYKNFFDHVDIQEKNINILNGNTEDHDAECQRYEEKIKSYGKIHLFMGGVGVDGHIAFNEPASSLSSRTRIKTLTEDTLIANSRFFDNDVNKVPKYALTIGVGTLLDAEEVMILVTGYNKAQALQAAVEGSINHLWTVTALQMHRRAIIVCDEPATQELKVKTVKYFTELEASAIRSVK"

linker = 'G' * 25

homodimer_sequence = sequence + linker + sequence
````

We tokenise the homodiemer sequence just like monomer above.

````python
tokenized_homodimer = tokenizer([homodimer_sequence], return_tensors="pt", add_special_tokens=False)
````

Now here's the tricky bit - we need to tweak the inputs so the model doesn't think this is just a single peptide. This is achieved by manipulating the position_ids input to the model. The position_ids input specifies the position of each amino acid in the input chain. By default, the model assumes you've passed a single, linear, contiguous chain. For example, if you provide a peptide with 100 amino acids, the model will assume the position_ids are simply [0, 1, ..., 98, 99] unless specified otherwise.

To clearly indicate that the two subunits aren't connected, we need to add a large offset to the position IDs of the second chain. The original repository uses an offset of 512, so we will adhere to that convention.

````python
with torch.no_grad():
    position_ids = torch.arange(len(homodimer_sequence), dtype=torch.long)
    position_ids[len(sequence) + len(linker):] += 512
print(position_ids)
````

Now we're ready to predict! Follow these steps to finalize the inputs:

1. Add position_ids: Incorporate the position_ids into the tokenized inputs.
2. Singleton Batch Dimension: Ensure to add a singleton batch dimension first to match the other arrays in the inputs.
3. Transfer to GPU: Once the inputs are prepared, transfer the dictionary to the GPU.
After completing these steps, we will be ready to obtain our folded structure.








````python
tokenized_homodimer['position_ids'] = position_ids.unsqueeze(0)

tokenized_homodimer = {key: tensor.cuda() for key, tensor in tokenized_homodimer.items()}
````

````python
with torch.no_grad():
    output = model(**tokenized_homodimer)
````

Next, we need to remove the poly-G linker from the output, so we can display the structure as fully independent chains. To do that, we'll alter the `atom37_atom_exists` field in the output.

The atom37_atom_exists field in the output of the ESMFold model is a PyTorch tensor that indicates whether each atom in the atom37 representation is present or not. The atom37 representation is a common way of representing the 3D structure of a protein, where each amino acid residue is represented by 37 atoms. The atom37_atom_exists tensor is a binary tensor of shape (num_residues, 37), where each element is 1 if the corresponding atom is present, and 0 if it is not.This information is useful for visualizing and interpreting the predicted protein structure, as it allows you to identify which atoms are confidently predicted versus which ones may be missing or uncertain.

 We will simply set all of the atoms for each of the linker residues to 0.

````python
linker_mask = torch.tensor([1] * len(sequence) + [0] * len(linker) + [1] * len(sequence))[None, :, None]

output['atom37_atom_exists'] = output['atom37_atom_exists'] * linker_mask.to(output['atom37_atom_exists'].device)
````

save the output to PDB file

````python
pdb = convert_outputs_to_pdb(output)
````

visualise

````python
view = py3Dmol.view(js='https://3dmol.org/build/3Dmol.js', width=800, height=400)
view.addModel("".join(pdb), 'pdb')

# The plddt field is scaled from 0-1 on earlier versions of ESMFold but will be updated
# to match AlphaFold's scale of 0-100 in future versions.
# We check here so that this code will work on either:

if torch.max(output['plddt']) <= 1.0:
    vmin = 0.5
    vmax = 0.95
else:
    vmin = 50
    vmax = 95

view.setStyle({'cartoon': {'colorscheme': {'prop':'b','gradient': 'roygb','min': vmin,'max': vmax}}})
````
