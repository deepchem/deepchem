#### Basic Featurization Tutorial

In this tutorial, we will learn about **featurizers** in DeepChem.

A featurizer converts raw data (like molecule names) into numbers
that machine learning models can understand.

We will cover 4 featurizers:
1. `Featurizer` — the base class
2. `OneHotFeaturizer` — converts categories to 0s and 1s
3. `RawFeaturizer` — passes molecule as-is
4. `UserDefinedFeaturizer` — use your own custom features

```python
#importing the libraries
import deepchem as dc
import numpy as np

print("DeepChem version:", dc.__version__)
```

### 1. OneHotFeaturizer

```python
#OneHot means: turn each category into a list of 0s and 1s
#Example: we have 3 atom types: Carbon(C), Nitrogen(N), Oxygen(O)
#C → [1, 0, 0]
#N → [0, 1, 0]
#O → [0, 0, 1]

featurizer = dc.feat.OneHotFeaturizer(
    allowable_set=["C", "N", "O", "F", "P"]
)

# Featurize some atoms
atoms = ["C", "N", "O", "F"]
result = featurizer.featurize(atoms)

print("Input atoms:", atoms)
print("\nOneHot output:")
for atom, features in zip(atoms, result):
    print(f"  {atom} → {features}")

print("\nShape of output:", result.shape)
print("Each atom becomes a vector of length:", result.shape[1])
```

##### What just happened?

Each atom got converted to a list of numbers:
- `C` → `[1, 0, 0, 0, 0]` means "it IS Carbon"
- `N` → `[0, 1, 0, 0, 0]` means "it IS Nitrogen"
- `F` → `[0, 0, 0, 1, 0]` means "it IS Fluorine"

This is called **One-Hot Encoding**. 
The machine learning model now understands these numbers!

### 2. RawFeaturizer

```python
""" RawFeaturizer is the simplest featurizer.
It just returns the molecule as a SMILES string
or as an RDKit molecule object — no transformation!
Useful when you want to pass raw molecules to a model."""

from rdkit import Chem

# Create some molecules using SMILES notation
# SMILES is a way to write molecules as text
# "C" = methane, "CCO" = ethanol, "c1ccccc1" = benzene
smiles_list = ["C", "CCO", "c1ccccc1"]

# RawFeaturizer with smiles=True returns SMILES strings
raw_featurizer = dc.feat.RawFeaturizer(smiles=True)
result = raw_featurizer.featurize(smiles_list)

print("Input SMILES:", smiles_list)
print("\nRawFeaturizer output (smiles=True):")
for smile, feat in zip(smiles_list, result):
    print(f"  {smile} → {feat}")
```

### 3.UserDefinedFeaturizer

```python
"""  Sometimes YOU already have features calculated
(like molecular weight, number of bonds, etc.)
UserDefinedFeaturizer lets you use YOUR OWN features! """

#For Example
# Let's say we have 3 molecules and
# we already calculated 2 features for each:
# [molecular_weight, number_of_rings]

import pandas as pd

# Our custom data
data = {
    "mol_id": ["mol1", "mol2", "mol3"],
    "smiles": ["C", "CCO", "c1ccccc1"],
    "mol_weight": [16.04, 46.07, 78.11],   # custom feature 1
    "num_rings": [0, 0, 1]                  # custom feature 2
}

df = pd.DataFrame(data)
print("Our custom molecule data:")
print(df)

# Now use UserDefinedFeaturizer to load these features
feature_columns = ["mol_weight", "num_rings"]
featurizer = dc.feat.UserDefinedFeaturizer(feature_columns)

# Featurize — pass the dataframe rows
features = featurizer.featurize(df)

print("\nUserDefinedFeaturizer output:")
for i, feat in enumerate(features):
    print(f"  mol{i+1} → {feat}")

print("\nShape:", features.shape)
print("Each molecule has", features.shape[1], "features")
```
